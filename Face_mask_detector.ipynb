```python
import cv2
```

```python
img=cv2.imread('pexels-cottonbro-3957987.jpg')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

```python
haar_cas=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
haar_cas.detectMultiScale(img)
```

```python
import numpy as np
```

```python
#this is for storing data without_mask
capture=cv2.VideoCapture(0) #0 for default camera
data=[] #create an empty array
while True:
    flag,img = capture.read()
    if flag:
        faces=haar_cas.detectMultiScale(img)
        for x,y,w,h in faces:
            cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,255),4)
            face=img[y:y+h,x:x+w,:]
            face=cv2.resize(face,(50,50))
            print(len(data))
            if len(data)<20:
                data.append(face) #store the face in the data array
        cv2.imshow('image',img)
        if cv2.waitKey(2)==27 or len(data)>=10:
            break
capture.release()
cv2.destroyAllWindows()
```

```python
np.save('without_mask.npy',data)
```

```python
plt.imshow(data[3])
```

```python
#wear mask to store data with_mask
capture=cv2.VideoCapture(0) #0 for default camera
data1=[] #create an empty array
while True:
    flag,img = capture.read()
    if flag:
        faces=haar_cas.detectMultiScale(img)
        for x,y,w,h in faces:
            cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,255),4)
            face=img[y:y+h,x:x+w,:]
            face=cv2.resize(face,(50,50))
            print(len(data1))
            if len(data)<20:
                data1.append(face) #store the face in the data array
        cv2.imshow('image',img)
        if cv2.waitKey(2)==27 or len(data1)>=10:
            break
capture.release()
cv2.destroyAllWindows()
```

```python
np.save('with_mask.npy',data1)
```
